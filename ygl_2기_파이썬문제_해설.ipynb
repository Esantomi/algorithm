## dictionary 구축하기 (for, if)
다음과 같은 리스트가 있을 때 각각의 요소의 개수를 value 값으로 갖는 딕셔너리를 만드세요.

book_title = ['great', 'expectations', 'the', 'adventures', 'of', 'sherlock', 'holmes', 'the', 'great', 'gasby', 'hamlet', 'adventures', 'of', 'huckleberry', 'fin']


예시 출력)<br>
{'great': 2, 'expectations': 1, 'the': 2, 'adventures': 2, 'of': 2, 'sherlock': 1, 'holmes': 1, 'gasby': 1, 'hamlet': 1, 'huckleberry': 1, 'fin': 1}

##### 위의 문제를 아래의 3가지 방법으로 딕셔너리를 각각 만들어 봅시다.
- for와 if만 사용
- count()를 사용
- get() 사용

````python
# for , if 만 사용
book_title = ['great', 'expectations', 'the', 'adventures', 'of', 'sherlock', 'holmes', 'the', 'great', 'gasby', 'hamlet', 'adventures', 'of', 'huckleberry', 'fin']
dict_book = {}
for book in book_title:
  if book in dict_book:
    dict_book[book] = dict_book[book] + 1
  else:
    dict_book[book] =1



print(dict_book)
````

````python
# for문과 count 만 사용
book_title = ['great', 'expectations', 'the', 'adventures', 'of', 'sherlock', 'holmes', 'the', 'great', 'gasby', 'hamlet', 'adventures', 'of', 'huckleberry', 'fin']
dict_book2 = {}
for book in book_title:
  dict_book2[book] = book_title.count(book)


print(dict_book2)
````

````python
# for문과 get()을 사용
book_title = ['great', 'expectations', 'the', 'adventures', 'of', 'sherlock', 'holmes', 'the', 'great', 'gasby', 'hamlet', 'adventures', 'of', 'huckleberry', 'fin']
dict_book3 = {}
for book in book_title:
  dict_book3[book] = dict_book3.get(book, 0) + 1



print(dict_book3)
````

````python
# 정민형님
book_title = ['great', 'expectations', 'the', 'adventures', 'of', 'sherlock', 'holmes', 'the', 'great', 'gasby', 'hamlet', 'adventures', 'of', 'huckleberry', 'fin']
dict_book2 = {}
book_set = set(book_title)
for i in book_set:
    dict_book2[i] = book_title.count(i)
print(dict_book2)
````

## 최고가와 최저가의 차이를 변동폭으로 정의할 때 (시작가 + 변동폭)이 최고가 보다 높을 경우 "상승장", 그렇지 않은 경우 "하락장" 문자열을 출력하세요.
```
Key Name         |  Description
---------------    ----------------------------
opening_price    |   최근 24시간 내 시작 거래금액
closing_price    |   최근 24시간 내 마지막 거래금액
min_price        |   최근 24시간 내 최저 거래금액
max_price        |   최근 24시간 내 최고 거래금액
```


````python
# 힌트
import requests
url = 'https://api.bithumb.com/public/ticker/btc'
data = { k: float(v) for k, v in requests.get(url).json()['data'].items() }

change_price = data['max_price'] - data['min_price']
if data['max_price'] < (data['opening_price'] + change_price):
  print("상승장")
else:
  print("하락장")

print("opening_price", data['opening_price'])
print("max_price", data['max_price'])
print("change_price", change_price)
print("today_price", data['opening_price']+change_price)
````

## 장바구니에 아래와 같은 과일이 들어있고 과일 판별 리스트가 있습니다. 현재 장바구니에는 과일이 몇개이고 과일이 아닌 것은 몇개인지 출력하세요.

```
basket_items = {'apples': 4, 'oranges': 19, 'kites': 3, 'sandwiches': 8}
fruits = ['apples', 'oranges', 'pears', 'peaches', 'grapes', 'bananas']
```

````python
'''
예시 출력)
과일은 23개이고, 11개는 과일이 아닙니다..
'''
basket_items = {'apples': 4, 'oranges': 19, 'kites': 3, 'sandwiches': 8}
fruits = ['apples', 'oranges', 'pears', 'peaches', 'grapes', 'bananas']
fruit = 0
not_fruit = 0

for item in basket_items:
  # print(item)
  if item in fruits:
    fruit += basket_items.get(item)
  else:
    not_fruit += basket_items.get(item)



print('과일은 {}개이고, {}개는 과일이 아닙니다.'.format(fruit, not_fruit))
````

## 영어 이름은 가운데 이름을 가지고 있는 경우가 있습니다. 가운데 이름은 대문자로 축약해서 나타내는 코드를 작성하세요.

````python
'''
예시 입력)
Alice Betty Catherine Davis
예시 출력)
Alice B. C. Davis
'''

def eng_name(fullname):
  fullname = fullname.split(' ')
  # print(fullname)
  result = ''
  for i in fullname:
    # print(i)
    if i == fullname[0] or i == fullname[-1]:
      result = result + ' ' + i
    else:
      i = i[0] + '.'
      result = result + ' ' + i
  
  print(result)

eng_name('Alice Betty Catherine Davis')
````

## 1월 1일 월요일부터 12월 31일까지 달력을 출력하세요.

````python
# 가능 #################
lst = [7,4,2,0,0,6,0,7,0, 0, 0]
# lst = [2,1,4,3,3,6,3,4,3]
# lst = [0,0,0,0,0,0,0,0,0]
# lst = [1,0,0,0,0,0,0,0,7]
# lst = [1,0,0,2,0,0,0,0,7]

# lst = list(map(int, input().split()))

loop = max(lst)
results = []
for height in range(loop):
    layer = [0 if i > height else 1 for i in lst]
    idx = layer.index(0)
    drop = layer[idx:]
    results.append(sum(drop))
print(max(results))
````

````python
# 가능 ##########################

# boxes = list(map(int, input().split()))
boxes = [3, 4, 2, 0, 0, 6, 0, 7, 0]
pile = [0] * max(boxes)
boxes.reverse()
result = 0

for i in range(len(boxes)):
    box = boxes[i]
    for j in range(box):
        if  i > pile[j]:
            result = max(result, i - pile[j])
        pile[j] += 1    

print(result)
````

````python
# X
x = [7, 4, 2, 0, 0, 6, 0, 7, 0]
max = 0
for n in range(9):
    x_temp = [int(i >= n+1) for i in x]
    x_temp = ''.join(map(str, x_temp))
    box_left = x_temp.find('1')
    box_shift_left = 9 - x_temp.count('1')

    shift = box_shift_left - box_left

    # print(f'{n} : {x_temp}')
    # print(f'\t{box_left} -> {box_shift_left} ({shift}만큼 이동)')

    if box_left == -1:
        continue
    
    if max < shift:
        max = shift

print(max)
````
